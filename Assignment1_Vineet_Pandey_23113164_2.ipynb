```python
# importing essential libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

plt.style.use('ggplot')
sns.set_palette('viridis')
```

```python
# dataset loading 
#dataset link : https://www.kaggle.com/datasets/ak0212/indian-car-market-dataset
df = pd.read_csv('car_dataset_india.csv')
df.head()
```

```python
df.columns # finding columns names
```

```python
#  cleaning of data and exploration
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.info()
df.isnull().sum()
df.describe()

```

```python
# Univariate Analysis  
# Distribution of Car Prices
plt.figure(figsize=(8, 5))
sns.histplot(df['price'], bins=20, kde=True, color='blue')
plt.title('Distribution of Car Prices')
plt.xlabel('Price (in INR)')
plt.show()
```

```python
# Count of Car Brands
plt.figure(figsize=(8, 5))
sns.countplot(x='brand', data=df, order=df['brand'].value_counts().head(10).index)
plt.title('Top 10 Car Brands in India')
plt.xticks(rotation=45)
plt.show()

```

```python
# Distribution of Car Mileage
plt.figure(figsize=(8, 5))
sns.histplot(df['mileage'], bins=20, kde=True, color='green')
plt.title('Distribution of Car Mileage')
plt.xlabel('Mileage (km/l)')
plt.show()
```

```python
# Count of Fuel Types
plt.figure(figsize=(8, 5))
sns.countplot(x='fuel_type', data=df)
plt.title('Count of Different Fuel Types')
plt.show()

```

```python
# Count of Transmission Types
plt.figure(figsize=(8, 5))
sns.countplot(x='transmission', data=df)
plt.title('Count of Transmission Types')
plt.show()
```

```python
# Multivariate Analysis  
# Correlation Heatmap
numerical_df = df.select_dtypes(include='number')

plt.figure(figsize=(10, 8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Average Price by Car Brand
plt.figure(figsize=(8, 5))
sns.barplot(x='brand', y='price', data=df, order=df['brand'].value_counts().head(10).index)
plt.title('Average Car Price by Brand')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average Price by Car Brand
plt.figure(figsize=(8, 5))
sns.barplot(x='brand', y='price', data=df, order=df['brand'].value_counts().head(10).index)
plt.title('Average Car Price by Brand')
plt.xticks(rotation=45)
plt.show()
```

```python
# Mileage vs Price
plt.figure(figsize=(8, 5))
sns.scatterplot(x='mileage', y='price', data=df)
plt.title('Mileage vs Car Price')
plt.show()
```

```python
# Engine Capacity vs Price
plt.figure(figsize=(8, 5))
sns.scatterplot(x='engine_cc', y='price', data=df)
plt.title('Engine Capacity vs Car Price')
plt.show()
```

```python
# Service Cost vs Price
plt.figure(figsize=(8, 5))
sns.scatterplot(x='service_cost', y='price', data=df)
plt.title('Service Cost vs Car Price')
plt.show()
```

```python
# Statistical Measures Calculations 
price_mean = df['price'].mean()
price_median = df['price'].median()
price_mode = df['price'].mode()[0]

print(f"Mean Car Price: {price_mean} INR")
print(f"Median Car Price: {price_median} INR")
print(f"Mode Car Price: {price_mode} INR")
```

```python
mileage_mean = df['mileage'].mean()
mileage_median = df['mileage'].median()
mileage_mode = df['mileage'].mode()[0]

print(f"Mean Mileage: {mileage_mean} km/l")
print(f"Median Mileage: {mileage_median} km/l")
print(f"Mode Mileage: {mileage_mode} km/l")
```

```python
#Box Plot : Price by Car Brand 
plt.figure(figsize=(10, 6))
sns.boxplot(x='brand', y='price', data=df, palette='viridis')
plt.title('Car Price by Brand')
plt.xticks(rotation=45)
plt.xlabel('Brand')
plt.ylabel('Price (INR)')
plt.show()

```

```python
#Engine Capacity vs Car Price
plt.figure(figsize=(8, 5))
sns.scatterplot(x='engine_cc', y='price', data=df, hue='fuel_type', palette='viridis')
plt.title('Engine Capacity vs Car Price')
plt.xlabel('Engine Capacity (CC)')
plt.ylabel('Price (INR)')
plt.show()
```

```python
# Pair Plots
sns.pairplot(numerical_df)
plt.show()
```

## 8. Conclusions  
- Popular brands like Maruti, Hyundai, and Tata dominate the Indian market.  
- Cars with larger engine capacity generally have higher prices.  
- Service cost increases with higher car prices, suggesting more expensive cars have higher maintenance.  
- Fuel efficiency (mileage) and price dont always correlate  luxury cars often prioritize power over efficiency.
